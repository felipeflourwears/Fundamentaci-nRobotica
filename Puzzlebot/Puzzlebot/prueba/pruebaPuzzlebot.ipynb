```python
def preprocessImage(img):
    # Your code here...
    width = int(img.shape[0])
    height = int(img.shape[1])
    img = cv2.resize(img,(height,width))
    img =cv2.GaussianBlur(img,(7,7),0)
    plt.figure(figsize=(10,10))
    plt.imshow(img,cmap="gray")

preprocessImage(imageRGB)
```
